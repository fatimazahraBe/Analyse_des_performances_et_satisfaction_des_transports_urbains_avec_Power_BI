```python
def rdf_daily_to_df(file_path):
    g = Graph()
    g.parse(file_path, format="xml")

    rows = {}

    for s, p, o in g:
        s = str(s)
        p = str(p)
        o = str(o)

        if s not in rows:
            rows[s] = {
                "subject": s,
                "city": "Chicago"
            }

        col = p.split("/")[-1]

        if col == "route":
            rows[s]["route"] = o
        elif col == "date":
            rows[s]["date"] = o
        elif col == "daytype":
            rows[s]["daytype"] = o
        elif col == "rides":
            rows[s]["ridership"] = o

    df = pd.DataFrame(rows.values())

    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    df["ridership"] = pd.to_numeric(df["ridership"], errors="coerce")

    return df
```

```python
folder_path = "/Workspace/Users/fatimazahrabellala56@gmail.com/datachicago"


dfs = []

for file in os.listdir(folder_path):
    if file.endswith(".rdf"):
        file_path = os.path.join(folder_path, file)
        print(f"Lecture du fichier : {file}")
        df_temp = rdf_daily_to_df(file_path)
        dfs.append(df_temp)
```

```python

df_Daily_All = pd.concat(dfs, ignore_index=True)
df_Daily_All.head()
```

```python
df_Daily_All = df_Daily_All.drop(columns=["city","subject"])
```
